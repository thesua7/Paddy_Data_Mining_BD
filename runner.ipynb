```python
data.hist(figsize=(14,12))
mplt.show()
```

```python
# Pair Plot
```

```python
data.reset_index(inplace=True)

sns.pairplot(data)
mplt.show()
```

```python
# Correlation Matrix
```

```python
size = 10
df = pd.DataFrame(data)
 
corrM = df.corr()
sns.heatmap(corrM, annot=True)
figure = mplt.gcf()
figure.set_size_inches(10,8)

mplt.show()
```
